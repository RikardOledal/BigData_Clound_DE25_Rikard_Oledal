```sql
%python
from pathlib import Path

volume_path = '/Volumes/data/nobel_prize/raw_data'
```

```sql
%python
from pyspark.sql.types import StructField, StringType, ShortType, IntegerType, FloatType, DateType, StructType

schema = StructType([
    StructField("awardYear", ShortType()),
    StructField("category", StringType()),
    StructField("categoryFullName", StringType()),
    StructField("sortOrder", ShortType()),
    StructField("portion", StringType()),
    StructField("prizeAmount", IntegerType()),
    StructField("prizeAmountAdjusted", IntegerType()),
    StructField("dateAwarded", DateType()),
    StructField("prizeStatus", StringType()),
    StructField("motivation", StringType()),
    StructField("categoryTopMotivation", StringType()),
    StructField("award_link", StringType()),
    StructField("id", ShortType()),
    StructField("name", StringType()),
    StructField("knownName", StringType()),
    StructField("givenName", StringType()),
    StructField("familyName", StringType()),
    StructField("fullName", StringType()),
    StructField("penName", StringType()),
    StructField("gender", StringType()),
    StructField("laureate_link", StringType()),
    StructField("birth_date", DateType()),
    StructField("birth_city", StringType()),
    StructField("birth_cityNow", StringType()),
    StructField("birth_continent", StringType()),
    StructField("birth_country", StringType()),
    StructField("birth_countryNow", StringType()),
    StructField("birth_locationString", StringType()),
    StructField("death_date", DateType()),
    StructField("death_city", StringType()),
    StructField("death_cityNow", StringType()),
    StructField("death_continent", StringType()),
    StructField("death_country", StringType()),
    StructField("death_countryNow", StringType()),
    StructField("death_locationString", StringType()),
    StructField("orgName", StringType()),
    StructField("nativeName", StringType()),
    StructField("acronym", StringType()),
    StructField("org_founded_date", DateType()),
    StructField("org_founded_city", StringType()),
    StructField("org_founded_cityNow", StringType()),
    StructField("org_founded_continent", StringType()),
    StructField("org_founded_country", StringType()),
    StructField("org_founded_countryNow", StringType()),
    StructField("org_founded_locationString", StringType()),
    StructField("ind_or_org", StringType()),
    StructField("residence_1", StringType()),
    StructField("residence_2", StringType()),
    StructField("affiliation_1", StringType()),
    StructField("affiliation_2", StringType()),
    StructField("affiliation_3", StringType()),
    StructField("affiliation_4", StringType()),
])

df_nobel = spark.read.csv(volume_path + "/complete.csv", header=True, schema=schema)

df_nobel.limit(10).display()
```

```sql
%python
df_nobel.orderBy('awardYear', 'category').display()
```

```sql
%python
df_nobel.createOrReplaceTempView("df_nobel")

spark.sql("""
        select
            birth_country,
            count(*) as count
        from df_nobel
        group by birth_country
        order by count desc
        """).display()
```
